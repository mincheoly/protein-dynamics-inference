```python
# Load cluster labels
import numpy as np

# Load isomap cluster labels
iso_cluster_30_40_97 = np.load('.//Sherlock_data//isomap_clustering_labels_30_40_97.dat')
iso_cluster_30_45_97 = np.load('.//Sherlock_data//isomap_clustering_labels_30_45_97.dat')
iso_cluster_35_40_97 = np.load('.//Sherlock_data//isomap_clustering_labels_35_40_97.dat')
iso_cluster_40_30_97 = np.load('.//Sherlock_data//isomap_clustering_labels_40_30_97.dat')
iso_cluster_list = [iso_cluster_30_40_97, iso_cluster_30_45_97, iso_cluster_35_40_97, iso_cluster_40_30_97]

# Load msm cluster labels
# Note: we turned off the ergodic cut off so there are 100 clusters
msm_cluster = np.load('.//Sherlock_data//msm_clustering_labels.dat')
msm_pop = np.load('.//Sherlock_data//msm_pop_vec.dat')
```

```python
# Calculate percentage of each msm cluster by counting

count = np.zeros([100], dtype=int)
for cluster in msm_cluster:
    count[cluster] += 1
msm_pop_est = count/float(sum(count))
print msm_pop_est
```

```python
# Calculate percentage of each isomap cluster by counting

iso_cluster = iso_cluster_list[0]
count = np.zeros([100], dtype=int)
for cluster in iso_cluster:
    count[cluster] += 1
iso_pop_est = count/float(sum(count))
print iso_pop_est
```

```python
#displays the distrubtion of the most popular msm cluster in isomap

cluster_id = np.argmax(msm_pop_est)
msm_frames = np.where(msm_cluster == cluster_id)[0]

iso_count = np.zeros([100], dtype=int)
for idx in msm_frames:
    iso_count[iso_cluster[idx]] += 1

print iso_count
print iso_count[np.argmax(iso_pop_est)] == max(iso_count), iso_count[np.argmax(iso_pop_est)]/float(sum(iso_count))
print "percent of msm frames that belong in the most represented isomap cluster"
```

```python
#displays the distrubtion of the most popular isomap cluster in msm space

cluster_id = np.argmax(iso_pop_est)
iso_frames = np.where(iso_cluster == cluster_id)[0]

msm_count = np.zeros([100], dtype=int)
for idx in iso_frames:
    msm_count[msm_cluster[idx]] += 1

print msm_count
print msm_count[np.argmax(msm_pop_est)]==max(msm_count), msm_count[np.argmax(msm_pop_est)]/float(sum(msm_count))
print "percent of isomap frames that belong in the most represented msm cluster"
```

```python
# Check if the largest n clusters in isomap and msm overlap the most with each other
# Only compares the nth largest isomap cluster with the nth largest msm cluster

#pick the iso_map to use
iso_cluster = iso_cluster_list[0]

# get population estimate for msm and isomap
count = np.zeros([100], dtype=int)
for cluster in msm_cluster:
    count[cluster] += 1
msm_pop_est = count/float(sum(count))

count = np.zeros([100], dtype=int)
for cluster in iso_cluster:
    count[cluster] += 1
iso_pop_est = count/float(sum(count))

# check overlap of the nth largest cluster for msm and isomap
for n in range(0,10):
    msm_cluster_id = msm_pop_est.argsort()[:][::-1][n]
    iso_cluster_id = iso_pop_est.argsort()[:][::-1][n]
    
    msm_frames = np.where(msm_cluster == msm_cluster_id)[0]
    iso_frames = np.where(iso_cluster == iso_cluster_id)[0]

    iso_count = np.zeros([100], dtype=int)
    for idx in msm_frames:
        iso_count[iso_cluster[idx]] += 1

    msm_count = np.zeros([100], dtype=int)
    for idx in iso_frames:
        msm_count[msm_cluster[idx]] += 1

    max_msm = msm_count[msm_cluster_id]
    max_iso = iso_count[iso_cluster_id]

    print "cluster " + str(n + 1) +":"
    print np.argmax(msm_count) == msm_cluster_id
    print "%.2f, %.2f" % (max_msm/float(sum(msm_count)), max_iso/float(sum(iso_count))) #percent in common
    print ""
```

```python
#find best match for each isomap cluster
#for each isomap cluster it looks for the best msm cluster that fits

#pick the iso_map to use
iso_cluster = iso_cluster_list[0]

# get population estimate for msm and isomap
count = np.zeros([100], dtype=int)
for cluster in msm_cluster:
    count[cluster] += 1
msm_pop_est = count/float(sum(count))

count = np.zeros([100], dtype=int)
for cluster in iso_cluster:
    count[cluster] += 1
iso_pop_est = count/float(sum(count))

print "isomap cluster, msm cluster"
# pick the isomap cluster
for m in range(0,10):
    iso_cluster_id = iso_pop_est.argsort()[:][::-1][m]
    iso_frames = np.where(iso_cluster == iso_cluster_id)[0]

    # check overlap of the nth largest cluster for msm and isomap
    for n in range(0,100):
        msm_cluster_id = msm_pop_est.argsort()[:][::-1][n]
        msm_frames = np.where(msm_cluster == msm_cluster_id)[0]

        msm_count = np.zeros([100], dtype=int)
        for idx in iso_frames:
            msm_count[msm_cluster[idx]] += 1
            
        iso_count = np.zeros([100], dtype=int)
        for idx in msm_frames:
            iso_count[iso_cluster[idx]] += 1
        
        max_iso = iso_count[iso_cluster_id]
        max_msm = msm_count[msm_cluster_id]

        if np.argmax(msm_count) == msm_cluster_id:
            print m + 1,n + 1
            print "%.3f, %.3f" % (max_iso/float(sum(iso_count)), max_msm/float(sum(msm_count))) #percent in common
            print ""
```

```python
#find best match for each isomap cluster
#for each isomap cluster it looks for the best msm cluster that fits

#pick the iso_map to use
iso_cluster = iso_cluster_list[0]

#initialize array to store chart data
data = []

# get population estimate for msm and isomap
count = np.zeros([100], dtype=int)
for cluster in msm_cluster:
    count[cluster] += 1
msm_pop_est = count/float(sum(count))

count = np.zeros([100], dtype=int)
for cluster in iso_cluster:
    count[cluster] += 1
iso_pop_est = count/float(sum(count))

# pick the isomap cluster
for m in range(0,100):
    iso_cluster_id = iso_pop_est.argsort()[:][::-1][m]
    iso_frames = np.where(iso_cluster == iso_cluster_id)[0]

    # check overlap of the nth largest cluster for msm and isomap
    for n in range(0,100):
        msm_cluster_id = msm_pop_est.argsort()[:][::-1][n]
        msm_frames = np.where(msm_cluster == msm_cluster_id)[0]

        msm_count = np.zeros([100], dtype=int)
        for idx in iso_frames:
            msm_count[msm_cluster[idx]] += 1
            
        iso_count = np.zeros([100], dtype=int)
        for idx in msm_frames:
            iso_count[iso_cluster[idx]] += 1
        
        max_iso = iso_count[iso_cluster_id]
        max_msm = msm_count[msm_cluster_id]

        if np.argmax(msm_count) == msm_cluster_id:
            percent_msm_in_iso = max_msm/float(sum(msm_count))
            percent_msm_in_iso = round(percent_msm_in_iso,3)
            percent_iso_in_msm = max_iso/float(sum(iso_count))
            percent_iso_in_msm= round(percent_iso_in_msm,3)
            
            #isomap cluster x, msm cluster y, % of isomap cluster in msm cluster, % of msm cluster in isomap cluster, 
            #frame count of isomap cluster x, frame count of msm cluster y, number of frames in common
            temp = [m + 1,n + 1, percent_iso_in_msm, percent_msm_in_iso, sum(msm_count), sum(iso_count), max_iso]
            data.append(temp)


np.set_printoptions(suppress=True)
label = ['iso #', 'msm #', '%iso->msm', '%msm->iso', 'total iso', 'total msm', 'common']
print label
print np.array(data)
```
