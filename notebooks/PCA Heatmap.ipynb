```python
# Pick out the hotspots

file_location = 'C:\\Users\\Christian\\Documents\\GitHub\\Protein_Dynamics_TJ\\MSM_Builder_Test\\PC_Data\\'
method_list = ['X_pc_isomap','X_pc_kernelPCA'][0]
parameter = '_45_45_'
ID = '14000'

fig, ax = plt.subplots()
try:
     pca = np.load(file_location + method + parameter + ID + '_4.dat')
except:
    print "ID" + " failed"
            
x = pca[:,0]
y = pca[:,1]
xedges, yedges = np.linspace(min(x), max(x), 80), np.linspace(min(y), max(y), 80)
hist, xedges, yedges = np.histogram2d(pca[:,0],pca[:,1],(xedges, yedges))
hist = np.log(hist)
xidx = np.clip(np.digitize(pca[:,0], xedges), 0, hist.shape[0]-1)
yidx = np.clip(np.digitize(pca[:,1], yedges), 0, hist.shape[1]-1)
c = hist[xidx, yidx]
plt.scatter(pca[:,0], pca[:,1], c=c, edgecolor='none')
cb = plt.colorbar()
cb.set_label('log10(N)')
        
plt.title(method+ID)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')

# Hot spots
cluster_identity = np.loadtxt(file_location + 'sample_cluster_assignment_' + ID + '.csv', dtype=int, delimiter=',')
hotspot_list = [(-.35, 0), (0.4, -.2)]
frame_idx = []

for hotspot in hotspot_list:
    d = np.linalg.norm(pca[:,0:2] - hotspot, axis=1)
    frame_idx.append(np.argmin(d))

# Show result
for frame in frame_idx:
    print "(Frame ID, Cluster ID): ", (frame, cluster_identity[frame])

# Save frames
from msmbuilder.example_datasets import FsPeptide
fs_peptide = FsPeptide()

X = np.loadtxt(file_location + 'raw_XYZ_' + ID + '.csv', delimiter=',')
our_traj = np.reshape(X[frame_idx], (len(frame_idx), len(X[0])/3, 3))
md_traj = md.Trajectory(our_traj, md.load(fs_peptide.data_dir + '/fs-peptide.pdb').topology)
md_traj.save_xtc(file_location + 'hotspot.xtc') # Saves an xtc trajectory file
```
